# Image to Pencil Sketch with Python

### steps :-
    1. Read the image
    2. Convert the image to grayscale
    3. Invert the grayscale image
    4. Blur the inverted grayscale image
    5. Mixing the grayscale image with the inverted blury image to create pencil sketch

## Imports

```python
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

## Read the image

```python
img = cv2.imread('dog.jpg')

# converting BGR to RGB 

color_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(color_img)

# image = cv2.imread('dog.jpg')
# cv2.imshow('Dog', image)
# cv2.waitKey(0)
```

## Converting the original image to greyscale

```python
gray_img = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')

# cv2.imshow('Gray_image', gray_img)
# cv2.waitKey(0)
```

## Inverting the grayscale 

```python
invert_img = 255 - gray_img
plt.imshow(invert_img, cmap='gray')

# cv2.imshow("Inverted", invert_img)
# cv2.waitKey(0)
```

## Bluring the inverted grayscale image

```python
blurred = cv2.GaussianBlur(invert_img, ksize=(25, 25), sigmaX=0, sigmaY=0)
plt.imshow(blurred, cmap='gray')

# cv2.imshow("blurred", blurred)
# cv2.waitKey(0)
```

## Converting to pencil sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=255.0)
plt.imshow(pencil_sketch, cmap='gray')

# cv2.imshow("pencil sketch", pencil_sketch)
# cv2.waitKey(0)
```
